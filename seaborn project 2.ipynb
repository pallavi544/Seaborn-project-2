## Seaborn Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Histogram 

```python
sns.histplot(data=df,x="BMI",kde=True,stat='frequency')
plt.show()
```

### Scatterplot

```python
sns.scatterplot(data=df,x='Age',y='BMI',hue='Gender')
plt.xlabel("Age")

plt.ylabel("BMI")

plt.title("Relationship between Age and BMI")
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Weight',hue='Gender',size='BMI',sizes=(50,100))
plt.xlabel("Age")

plt.ylabel("Weight")

plt.title("Relationship between Age and Weight")
plt.show()
```

### Barplot

```python
sns.barplot(y='Age',x='Label',data=df,hue='Gender')
plt.xlabel("label")
plt.ylabel("Age")
plt.title("Relationship between Age and Label")
plt.show()
```

```python
sns.barplot(y='BMI',x='Gender',data=df,color='pink')
plt.xlabel("Gender")
plt.ylabel("BMI")
plt.title("Relationship between Gender and BMI")
plt.show()
```

```python
sns.barplot(y='Weight',x='Gender',data=df,color='skyblue')
plt.xlabel("Gender")
plt.ylabel("Weight")
plt.title("Relationship between Gender and weight")
plt.show()
```

```python
sns.barplot(y='Height',x='Gender',data=df,color='magenta')
plt.xlabel("Gender")
plt.ylabel("Height")
plt.title("Relationship between Gender and Height")
plt.show()
```

### stirplot and Swarmplot

```python
sns.stripplot(y='Age',x='Label',data=df,hue='Gender')
plt.show()
```

```python
sns.swarmplot(y='BMI',x='Gender',data=df,color='pink')
plt.show()
```

### Countplot

```python
sns.countplot(x='Gender',data=df)
plt.show()
```

```python
sns.countplot(x='Label',data=df,color="green")
plt.show()
```

### Boxplot and Violinplot

```python
sns.boxplot(x='Label',y='Age',hue='Gender',data=df)
plt.show()
```

```python
sns.boxplot(y='BMI',x='Gender',data=df)
plt.show()
```

```python
sns.violinplot(x='Label',y='Age',data=df)
plt.show()
```

### KDEplot and Regression plot

```python
sns.kdeplot(data=df,x='Age',y='BMI',shade=True,color='red')
```

```python
sns.regplot(data=df,x='Age',y='BMI')
```

### Pairplot and Jointplot

```python
sns.pairplot(df)
```

```python
sns.jointplot(data=df,x='Age',y='BMI',kind='scatter')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in ['Gender','Label']:
    df[column]=le.fit_transform(df[column])
```
